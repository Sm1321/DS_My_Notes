```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris.feature_names)
```

```python
print(iris.target_names)
```

```python
print(iris['target'])
```

```python
## independent features
X = pd.DataFrame(iris['data'],columns = iris['feature_names'])
```

```python
### Y dependent features

y = iris['target']
```

```python
X.head(15)
```

```python
### Train,test,split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 40)
```

```python
print("shapes of X:-",X_train.shape," ",X_test.shape)
print("shapes of y:-",y_train.shape," ",y_test.shape)
```

```python
## Apply Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()
```

```python
tree_clf.fit(X_train,y_train)
```

```python
X_train.head(10)
```

```python
(tree_clf)
```

```python
### Visualize the Decsion tree

from sklearn import tree

plt.figure(figsize = (10,10))

tree.plot_tree(tree_clf,filled = True)

plt.show()
```

```python
## Apply Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(max_depth=2)
```

```python
tree_clf.fit(X_train,y_train)
```

```python
### Visualize the Decsion tree

from sklearn import tree

plt.figure(figsize = (10,10))

tree.plot_tree(tree_clf,filled = True)

plt.show()
```

```python
y_pred = tree_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test,y_pred)

print(cm)


print('--'*60)
print(classification_report(y_test,y_pred))
```

### Decison Tree Classifier Pre-Pruning and HyperParamater Tuning

```python
param = {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV

import warnings
warnings.filterwarnings('ignore')
```

```python
treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(treemodel,param_grid=param,cv = 5,scoring ='accuracy')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test,y_pred)

print("COnfusion MAtrix:-",cm)


print('--'*60)
print(classification_report(y_test,y_pred))
```

```python
score = accuracy_score(y_pred,y_test)

print("Accuracy Score:-",score)
```
