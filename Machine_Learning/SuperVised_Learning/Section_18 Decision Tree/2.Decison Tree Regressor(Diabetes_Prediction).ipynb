### **Diabetes Prediction**

```python
from sklearn.datasets import load_diabetes
```

```python
dataset  = load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
diabetes_df = pd.DataFrame(dataset.data,columns = dataset.feature_names)
```

```python
diabetes_df .head()
```

```python
## Independent and DEpendent FEatures
X = diabetes_df

y = dataset.target
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
### Correlation
X_train.corr()
```

```python
import seaborn as sns
plt.figure(figsize = (15,10))

sns.heatmap(X_train.corr(),annot = True)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

### **HyperParameter Tuning**

```python
param = {
    'criterion':['squared_error','fried_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
```

```python
grid= GridSearchCV(regressor,param_grid =param,cv = 5,scoring = 'neg_mean_squared_error')

```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
## Predictions
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print("R2 Score:-",r2_score(y_test,y_pred))
print("Mean Absolute Error:-",mean_absolute_error(y_test,y_pred))
print("Mean Squared Error:-",mean_squared_error(y_test,y_pred))
```

```python
selectedmodel = DecisionTreeRegressor(criterion='friedman_mse', max_depth=4, max_features='log2',splitter = 'random')
```

```python
selectedmodel.fit(X_train,y_train)
```

```python
 ## Vuisualize the DT Tree

import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(selectedmodel,filled = True)
```
