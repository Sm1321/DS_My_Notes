```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
### Create a dataset
X,y = make_classification(n_samples = 1000,n_features=10,n_classes=2,random_state=15)
```

```python
X
```

```python

y
```

```python
pd.DataFrame(X)
```

```python
print(X.shape," ",y.shape)
```

```python
from sklearn.model_selection import train_test_split # Correct the typo in the module name

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
## Model training
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train,y_train)

```

```python
y_pred = logistic.predict(X_test)
print(y_pred)
```

```python
logistic.predict_proba(X_test)
```

```python
#performace metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = (accuracy_score(y_test,y_pred))
print("Accuracy Score:-",score)
print("------------------------------------")
cm = (confusion_matrix(y_test,y_pred))
print("Confusion Matrix:-",cm)
print("------------------------------------")
print(classification_report(y_test,y_pred))
```

## **Hyperparameter Tuning And Cross Validation**

### **Grid Search CV**

```python
model = LogisticRegression()
penalty = ['l1','l2','elasticnet']
c_values = [100,10,1.0,0.1,0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
## put them in key values pairs
params = dict(penalty=penalty,C=c_values,solver=solver)
```

```python
params
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
## Grid Search Cv
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator = model,param_grid = params,scoring = 'accuracy',cv = cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
y_pred
```

```python
score = (accuracy_score(y_test,y_pred))
print("Accuracy Score:-",score)
print("------------------------------------")
cm = (confusion_matrix(y_test,y_pred))
print("Confusion Matrix:-",cm)
print("------------------------------------")
print(classification_report(y_test,y_pred))
```

### **Randomized SearchCV**

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
## put them in key values pairs
params = dict(penalty=penalty,C=c_values,solver=solver)
```

```python
model = LogisticRegression()
randomcv = RandomizedSearchCV(estimator = model,param_distributions = params,scoring = 'accuracy',cv = 5,n_jobs=-1)
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
 y_pred = randomcv.predict(X_test)
```

```python
y_pred
```

```python
score = (accuracy_score(y_test,y_pred))
print("Accuracy Score:-",score)
print("------------------------------------")
cm = (confusion_matrix(y_test,y_pred))
print("Confusion Matrix:-",cm)
print("------------------------------------")
print(classification_report(y_test,y_pred))
```
