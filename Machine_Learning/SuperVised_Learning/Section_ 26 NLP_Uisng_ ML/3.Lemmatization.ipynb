## **Wordnet Lemmatizer:-**

Lemmatization technique is like stemming . The Output we will get after Lemmatization is called 'lemma'.Which is a root word rather than root stem.


The Output of stemming ,After Lemmatization, we will be getting a valid word that means the same thing


NLTK ProvidesWordNetLemmatizer classwhich is a thin wrapper around the wordnet corpus .This Classes uses **morphy()** Function to **WordNet CorpusReader class** to find lemma.


-->This process will take more times

Let's Understend it with an Example:-

```python
import nltk
nltk.download('wordnet')
```

```python

## Q&A ,chatbots ,text summarization
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
'''
Pos Noun - n
verb -v
adjective -a
adverb -r
'''

lemmatizer.lemmatize("going",pos = 'v')
```

```python


from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

'''
Pos Noun - n
verb -v
adjective -a
adverb -r
'''

lemmatizer.lemmatize("going",pos = 'v')
```

```python

words = ["writing","writes","programming","programs","eating","eats","eaten","finally","finalized","history"]

```

```python
for word in words:
  print(word+"--->"+lemmatizer.lemmatize(word,pos='n'))
```

```python
for word in words:
  print(word+"--->"+lemmatizer.lemmatize(word,pos='v'))
```
