### **Stemming**

Stemming is the Process of reducing a word to its word stem that affixes to suffixes and pre-fixes or to the roots of words know as lemma.


Stemming is important in natural language Understanding (NLU) and Natural Language Processing(NLP)

```python
### Classifiaction  Problem


## Comments of product is a Positive review Or Negative Review


## Reviews --> eating , eat, eaten [ going, gone, goes] -> go


words = ["writing","writes","programming","programs","eating","eats","eaten","finally","finalized","history"]

```

### **PorterStemmer**

```python
from nltk.stem import PorterStemmer
```

**The PorterStemmer** is a class in the Natural Language Toolkit (NLTK) library used for stemming words. Stemming is the process of reducing a word to its base or root form.


 For example, the words "running", "runs", and "ran" can all be reduced to the root form "run".



 This is particularly useful in natural language processing tasks where you want to treat different forms of a word as the same term.

```python
stemming  = PorterStemmer()
```

```python
for word in words:
  print(word+"--->"+stemming.stem(word))
```

```python
stemming.stem('Congratulations')
```

```python
stemming.stem("sitting")
```

**RegexpStemmer Class**

NLTK has RegexpStemmer class with the help of which we can easily implement Regular EXpression Stemmer Algorithms.


It Basically Takes a single Regular Expression and Removes any Prefix (or) Suffix that Matches the Expression .


Let us see the Example:-

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer(r'ing$|e$|e$|able$',min = 4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem( 'eatingly')
```

```python
reg_stemmer.stem('ingeating')
```
